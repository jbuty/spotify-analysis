# Playing with My Spotify Data

## Using Spotipy

Spotipy is a lightweight Python Library for the Spotify Web API. I'm specifically using it to garner data about how my music listening habits have changed over time based on a number of audio statistics. Additionally, I can see the top tracks and artists that I've listened to the most for three different time frames. 

The Spotify API gives users access to their music listening data for the short-term ( ~4 weeks), medium-term ( ~6 months) and long-term time frames (several years). 


### Import Libaries

```python
# Initial Set Up
import csv
from ast import literal_eval

# Data Manipulation
import pandas as pd
import numpy as np

# For Visualizations
import matplotlib.pyplot as plt
import seaborn as sns
```

## Reading in my top track and top artist CSV Files


```python
# Reading in my top artist data
short_term_top = pd.read_csv('spotify_data/st_artists.csv', converters = {'genres': literal_eval})
medium_term_top = pd.read_csv('spotify_data/mt_artists.csv', converters = {'genres': literal_eval})
long_term_top = pd.read_csv('spotify_data/lt_artists.csv', converters = {'genres': literal_eval})

# Reading in my top track data
st_top_tracks_df = pd.read_csv('spotify_data/st_tracks.csv')
mt_top_tracks_df = pd.read_csv('spotify_data/mt_tracks.csv')
lt_top_tracks_df = pd.read_csv('spotify_data/lt_tracks.csv')
```

### Displaying Top Artists for Each Duration (Short, Medium, Long)

Sorting by popularity allows me to see which songs and artists I listen to are most popular among the Spotify community. The popularity score is calculated by an algorithm that takes into consideration not only number of plays but also how recent the plays occurred. Therefore, newer songs may be more scored a high popularity score even if they have smaller play count compared to old ones.

```python
# Sorting by popularity and subsetting important fields 
st_df = short_term_top[['name', 'popularity', 'genres', 'followers.total', 'uri']].sort_values(
    by = ['popularity'], ascending = False)

mt_df = medium_term_top[['name', 'popularity', 'genres', 'followers.total', 'uri']].sort_values(
    by = ['popularity'], ascending = False)

lt_df = long_term_top[['name', 'popularity', 'genres', 'followers.total', 'uri']].sort_values(
    by = ['popularity'], ascending = False)
```

## Get Audio Statistics for top songs for each Time Frame

```python
results_analytics = pd.read_csv('spotify_data/results_analytics.csv')
results_analytics = results_analytics.drop(['Unnamed: 0'], axis = 1)

results_analytics.index = ['Loudness (dB)', 'Tempo (BPM)', 'Modality', 'Danceability', 'Energy', 'Speechiness', 
                           'Acousticness', 'Instrumentalness', 'Liveness', 'Valence', 'Popularity']
results_analytics
```

Here we can compare different audio statistics for each time frame. While a lot of the audio statistics are fairly close in score across each time frame there are a few notable trends. Starting at **loudness**, I've started to listen to songs that are louder. The **tempo** of songs is decreasing which makes sense since historically I listen to a lot of up-beat house music but more recently have been listening to indie and down tempo EDM (electronic dance music). Additionally, we can see that my more recent music listening habits are increasingly **acoustic** and **live** while decreasing in **instrumentalness**. Instrumentalness is defined as whether a track contains no vocals. Liveness is aptly defined as the presence of an audience in the recording. Overall, I can see that my music listening habits are indeed changing as I am listening to more music from bands and individual singers compared to DJs. The average audio statistics for each time frame confirm my assumptions. Lastly, the popularity of a track is significantly for the long-term compared to medium and short term. I believe this has to do with the fact that older songs are played less often compared to newer ones.

For more information on the definition of each Spotify audio statistic check out this [link](https://developer.spotify.com/documentation/web-api/reference/tracks/get-audio-features/).

## Using Seaborn for visualizations

```python
top_songs = pd.read_csv('spotify_data/top_songs.csv')
```

```python
top_songs = top_songs.sort_values(by = ['popularity'], ascending = False)

top_songs_l = top_songs[top_songs['term'] == 'Long Term']
top_songs_m = top_songs[top_songs['term'] == 'Medium Term']
top_songs_s = top_songs[top_songs['term'] == 'Short Term']

popularity = top_songs['popularity']
artists = top_songs['artists']


plt.figure(figsize=(10,6))

sns.boxplot(x = top_songs['popularity'], y = top_songs['term'], data = top_songs)
plt.xlabel('Popularity (0-100)', fontsize = 14)
plt.ylabel('')
plt.title('Song Popularity by Time Frame', fontweight = 'bold', fontsize = 18)
```

Above, we can see my song popularity distribution for each time frame. The majority of songs for each time frame cluster between the high 30s and low 60s. However, for the long-term time frame the distribution is much much larger, starting at a score of 0 and stretching to high 70s. Next, I'll look at the song popularity distribution for each artist that I listen too. 

```python

fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize = (17,13))
fig.subplots_adjust(wspace = .36)
fig.suptitle('Artist Popularity Distribution By Time Frame', fontweight = 'bold', fontsize = 20)
plt.setp((ax1, ax2, ax3), xlim = (-5,90))

sns.boxplot(x = top_songs_l['popularity'], y = top_songs_l['artists'], data = top_songs_m, ax = ax1)
ax1.set_ylabel('Artists', fontsize = 14, fontweight = 'bold')
ax1.set_xlabel('')
ax1.set_title('Long Term (> 1 year)')

sns.boxplot(x = top_songs_m['popularity'], y = top_songs_m['artists'], data = top_songs_m, ax = ax2)
ax2.set(ylabel = '')
ax2.set_xlabel('Popularity (0-100)', fontsize = 14, fontweight = 'bold')
ax2.set_title('Medium Term (~ 6 Months)')

sns.boxplot(x = top_songs_s['popularity'], y = top_songs_s['artists'], data = top_songs_s, ax = ax3)
ax3.set(ylabel = '')
ax3.set_label('')
ax3.set_title('Short Term (~ 4 Weeks)')
```

While these graphs are fairly large, I still wanted to include all artists for each time frame. One thing to note, Spotify only returns the top 50 tracks for each time frame so it's likely that I only have one track returned for my top artists since I listen to a wide spectrum of artists. For the long term, I have a number of artists with only one track and a score of 0. This explains the much larger distribution for song popularity compared to the other time frames. Across all 3 times frames, it's pretty clear that only a handful of artists have more than one song recorded in my listening habits. I seldom listen to albums from beginning to end and instead rely on playlists curated by me or others. The main exception here is Tycho as I listen to his music religiously. Additionally, in the short term, I've been listening to albums from Fleet Foxes, Leon Bridges, and Frank Ocean which is why they have a larger distribution of songs.
